## Identify extreme scores

```python
# Define the SQL query
query = """
SELECT
    e.Artist_ID,
    Name,
    Review,
    Sentiment_Score
FROM Events e
JOIN Reviews r ON e.Artist_ID = r.Artist_ID
WHERE Sentiment_Score > 90 OR Sentiment_Score < 10
ORDER BY Sentiment_Score DESC;
"""

# Execute the query and return a DataFrame
df = pd.read_sql(query, con=engine)

# Display the DataFrame
df
```

## Frequency of Different Emotions for Each Artist at Each Venue

```python
# Define the SQL query
query = """
WITH EmotionCounts AS (
    SELECT
        e.Venue_Name,
        e.Artist_ID,
        r.Emotion,
        COUNT(*) AS Emotion_Count
    FROM Events e
    JOIN Reviews r ON e.Artist_ID = r.Artist_ID
    GROUP BY e.Venue_Name, e.Artist_ID, r.Emotion
),
EmotionRanks AS (
    SELECT
        Venue_Name,
        Artist_ID,
        Emotion,
        Emotion_Count,
        RANK() OVER (PARTITION BY Venue_Name, Artist_ID ORDER BY Emotion_Count DESC) AS Emotion_Rank
    FROM EmotionCounts
)
SELECT c.Venue_Name, c.Artist_ID, c.Emotion, c.Emotion_Count
FROM EmotionRanks c
WHERE c.Emotion_Rank = 1;
"""

# Execute the query and return a DataFrame
df = pd.read_sql(query, con=engine)

# Display the DataFrame
df
```

## Ranking of Venues by Review Count and Sentiment Score

```python
# Define the SQL query
query = """
WITH VenueMetrics AS (
    SELECT
        e.Venue_Name,
        COUNT(r.Review) AS Total_Reviews,
        AVG(r.Sentiment_Score) AS Average_Sentiment
    FROM Events e
    JOIN Reviews r ON e.Artist_ID = r.Artist_ID
    GROUP BY e.Venue_Name
),
RankedVenues AS (
    SELECT
        Venue_Name,
        Total_Reviews,
        Average_Sentiment,
        RANK() OVER (ORDER BY Total_Reviews DESC) AS Review_Rank,
        RANK() OVER (ORDER BY Average_Sentiment DESC) AS Sentiment_Rank
    FROM VenueMetrics
)
SELECT
    Venue_Name,
    Total_Reviews,
    Average_Sentiment,
    Review_Rank,
    Sentiment_Rank
FROM RankedVenues
WHERE Review_Rank <= 5 OR Sentiment_Rank <= 5;
"""

# Execute the query and return a DataFrame
df = pd.read_sql(query, con=engine)

# Display the DataFrame
df
```
